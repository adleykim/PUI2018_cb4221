```python
# drop all columns except for zip code and land use

bk_2 = bk[['ZipCode', 'LandUse']]
bk_2.head(5)
```

```python
mh_2 = mh[['ZipCode', 'LandUse']]
mh_2.head(5)
```

```python
# removing NaN values (since this is a problem later)
bk_3 = bk_2.dropna()
```

```python
mh_3 = mh_2.dropna()
```

```python
bk_3.head()
```

```python
# grouping by landuse and then counting, then resetting index

bk_4 = bk_3.groupby(by='LandUse').count()
#print(bk_4)
bk_5 = bk_4.reset_index()
```

```python
# grouping by landuse and then counting, then resetting index

mh_4 = mh_3.groupby(by='LandUse').count()
#print(mh_4)
mh_5 = mh_4.reset_index()
```

```python
# merging two dfs with landuse

data = pd.merge(pd.DataFrame(bk_5), pd.DataFrame(mh_5), on='LandUse', left_index=True, right_index=True)
```

```python
data.columns
```

```python
# rearranging columns
data2 = data[['LandUse', 'ZipCode_x', 'ZipCode_y']]
```

```python
# renaming columns
data2.columns = ['LandUse', 'Brooklyn', 'Manhattan']  
```

```python
data2
```

```python
# not sure how to replace '01' with what I want, so I created a whole new column

data3=data2.assign(Landuse = ['One&Two Family Bldgs', 'Multi-Family Walk Up', 'Multi Family Elevator',
                            'Mixed Residential & Commercial', 'Commercial and Office', 'Industrial & Manufacturing',
                            'Transportation & Utility', 'Public Facilities & Institutions', 'Open Space & Rec',
                            'Parking Facilities', 'Vacant Land'])
data3
```

```python
data4=data3[['Landuse', 'Brooklyn', 'Manhattan']]
data4
```

```python
# setting up new column with manhattan to brooklyn ratio 

data4['luratio'] = data4.apply(lambda row: row.Manhattan / row.Brooklyn, axis=1)
data4
```

```python
# creating new column for ratio > 1; true/false
# search for: create new column in pandas based on a conditional 
data4['Manhattan Wins'] = np.where(data4['luratio']>1, 'True', 'False')
data4
```

The result above does make sense simply because of how much larger Brooklyn is compared to Manhattan (70 square miles versus 20 square miles, respectively.) Given the limited space in Manhattan, one would also expect to have a higher ratio of buildings with elevators, because density is achieved by building larger structures (the only place to go is up). Finally, given the premium on space in Manhattan, low ratios for parking facilities, industrial & manufacturing, and vacant land also makes sense.
